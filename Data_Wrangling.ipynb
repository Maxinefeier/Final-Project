```python
Annual_Gross1=pd.melt(Annual_Gross,id_vars=['Region'])
Annual_Gross1
Annual_Gross1=Annual_Gross1.rename(columns={'variable':'Year','value':'Gross Product'})
```

```python
AverageWage1=pd.melt(AverageWage,id_vars=['Region'])
AverageWage1=AverageWage1.rename(columns={'variable':'Year','value':'Average Wage'})
AverageWage1
```

```python
Budgetary_Expenditure1=pd.melt(Budgetary_Expenditure,id_vars=['Region'])
Budgetary_Expenditure1=Budgetary_Expenditure1.rename(columns={'variable':'Year','value':'Budgetary Expenditure'})
Budgetary_Expenditure1
```

```python
Urban_Area1=pd.melt(Urban_Area,id_vars=['Region'])
Urban_Area1=Urban_Area1.rename(columns={'variable':'Year','value':'Urban Area'})
Urban_Area1
```

```python
N_Cities1=pd.melt(N_Cities,id_vars=['Region'])
N_Cities1=N_Cities1.rename(columns={'variable':'Year','value':'Number of Cities'})
N_Cities1
```

```python
Unemployment_Rate1=pd.melt(Unemployment_Rate,id_vars=['Region'])
Unemployment_Rate1=Unemployment_Rate1.rename(columns={'variable':'Year','value':'Unemployment Rate'})
Unemployment_Rate1
```

```python
residece_population1=pd.melt(residece_population,id_vars=['Region'])
residece_population1=residece_population1.rename(columns={'variable':'Year','value':'Population'})
residece_population1
```

```python
N_Employed_Persons1=pd.melt(N_Employed_Persons,id_vars=['Region'])
N_Employed_Persons1=N_Employed_Persons1.rename(columns={'variable':'Year','value':'Number of Employed Population'})
N_Employed_Persons1
```

```python
df1=Annual_Gross1.merge(IndustrialEnterprises1,how="outer",on=["Region","Year"])
df1
```

```python
df2=df1.merge(AverageWage1,how="outer",on=["Region","Year"])
df2
```

```python
df3=df2.merge(Budgetary_Expenditure1,how="outer",on=["Region","Year"])
df4=df3.merge(Population_Density1,how="outer",on=["Region","Year"])
df5=df4.merge(Urban_Area1,how="outer",on=["Region","Year"])
df6=df5.merge(N_Cities1,how="outer",on=["Region","Year"])
df7=df6.merge(Unemployment_Rate1,how="outer",on=["Region","Year"])
df8=df7.merge(residece_population1,how="outer",on=["Region","Year"])
df9=df8.merge(N_Employed_Persons1,how="outer",on=["Region","Year"])
df9
```
